```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("House Price India.csv")
```

```python
df
```

```python
df.dtypes
```

**UNIVARIATE ANALYSIS**

**HISTOGRAM**

```python
cols = 5
rows = 5
num_cols = df.select_dtypes(exclude='object').columns
fig = plt.figure( figsize=(cols*4, rows*4))
for i, col in enumerate(num_cols):
    
    ax=fig.add_subplot(rows,cols,i+1)
    
    sns.histplot(x = df[col], ax = ax)
    
fig.tight_layout()  
plt.show()
```

KDE PLOT

```python
sns.kdeplot(x='Price',data=df)
```

BOX PLOT

```python
cols = 5
rows = 5
num_cols = df.select_dtypes(exclude='object').columns
fig = plt.figure(figsize= (15,9))
for i, col in enumerate(num_cols):
    
    ax=fig.add_subplot( rows, cols, i+1)
    
    sns.boxplot(x=df[col], ax=ax)
    
fig.tight_layout()  
plt.show()
```

**BIVARIATE**

```python
plt.scatter(df['living area'], df['Price'])
plt.title('Living Area  vs. Price')
plt.xlabel(' Living Area')
plt.ylabel('Price')
```

```python
plt.scatter(df['Price'], df['number of bedrooms'])
plt.title('Price  vs. Number of bedrooms')
plt.xlabel('Price')
plt.ylabel('Number of bedrooms')
```

```python
plt.scatter(df['Price'], df['Built Year'])
plt.title('Price  vs. Number of bedrooms')
plt.xlabel('Price')
plt.ylabel('Number of bedrooms')
```

**MULTIVARIATE**

```python
sns.pairplot(data=df[['number of bedrooms','living area','lot area','Built Year','Price']])
```
